# Some best practices

```python
# Use not 0,1,2,3, ... for row indexes
df1 = pd.read_csv(dataset_name,index_col="Title")
df1
```

```python
# Then use lower cases all around
```

```python
df1.columns = ["rank","genre","description","director","actors","year","length","rating","votes","revenues","metascore"]
df1.index.name = "title"
df1
```

```python
# Statistiche
df1.describe()
```

```python
# If I need to know how many null / missing values are present
df1.isnull().sum()
```

```python
# now we can fill the missing values with the average values
df1["revenues"].fillna( df1["revenues"].mean(), inplace=True)
df1["metascore"].fillna( df1["metascore"].mean(), inplace=True)
```

```python
df1.isnull().sum()
```

```python
# Let's remove some data (the entire 'rank' column)
# Note: axis==0 --> rows, axis==1 --> columns)
df1.drop(["rank"],axis=1, inplace=True)
```

```python
df1.tail()
```

# Searching and sorting

```python
# We can also perform aggregate statistics by some given column
df1.groupby(["year"]).mean()
```

```python
# Or who has made the most money
df1.groupby(["year"]).mean().sort_values(by=["revenues"], ascending=False)
```

```python
# Filter by condition too can 
df1[df1["year"]==2010].sort_values(by=["revenues"],ascending=False)
```

```python
desired_genre = ["Comedy","Action"]
df1[df1["genre"].isin(desired_genre)].sort_values(by=["rating"],ascending=False)
```

```python
# The best rated movie but with the least amount of revenues!
# We use 'quantile' to select all rows that are are in the 1% for rating but in the worst 25% for revenues
df1[
    (df1["rating"] > df1["rating"].quantile(0.99))
    & (df1["revenues"] < df1["revenues"].quantile(0.25))
]
```
